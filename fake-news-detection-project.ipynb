```python
import pandas as pd

# Load both datasets (make sure the files exist in your working directory)
df_fake = pd.read_csv("/kaggle/input/fake-and-true-datasets/Fake.csv")
df_true = pd.read_csv("/kaggle/input/fake-and-true-datasets/True.csv")

# Add labels
df_fake["class"] = 0  # Fake news
df_true["class"] = 1  # True news

# Combine datasets
df = pd.concat([df_fake, df_true], ignore_index=True)

# Check the class distribution
print(df["class"].value_counts())  # ✅ You should now see both 0 and 1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Split features and labels
x = df["text"]
y = df["class"]

# Stratified train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y, random_state=42)

# TF-IDF Vectorization
vectorizer = TfidfVectorizer()
xv_train = vectorizer.fit_transform(x_train)
xv_test = vectorizer.transform(x_test)

# Train model
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

# Step 1: Clean text
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>+', '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text

# Step 2: Sample Data
data = {
    'text': [
        "NASA finds life on Mars!",
        "Aliens invade Earth, government hides truth",
        "Apple launches new iPhone",
        "Click here for free money!!!"
    ],
    'class': [1, 0, 1, 0]  # 1 = Real, 0 = Fake
}
df = pd.DataFrame(data)
df['text'] = df['text'].apply(clean_text)

# Step 3: Train model
vectorizer = TfidfVectorizer()
x = vectorizer.fit_transform(df['text'])
y = df['class']
model = LogisticRegression()
model.fit(x, y)

# Step 4: Manual Test Function
def manual_testing(news):
    cleaned = clean_text(news)
    vec = vectorizer.transform([cleaned])
    pred = model.predict(vec)[0]
    print("✅ Result:", "Not A Fake News" if pred == 1 else "❌ Fake News")

# Step 5: Run test
test_news = """21st Century Wire says This week, the historic international Iranian Nuclear Deal was punctuated by a two-way prisoner swap between Washington and Tehran..."""
manual_testing(test_news)
```
